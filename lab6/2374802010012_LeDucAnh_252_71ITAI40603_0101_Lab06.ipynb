```python
import cv2
import matplotlib.pyplot as LeDucAnh
img = cv2.imread('image.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
face_data = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_data.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(rgb, (x,y), (x+w,y+h), (0,255,0), 3)
LeDucAnh.imshow(rgb)
LeDucAnh.title("Test Detection")
LeDucAnh.axis('off')
LeDucAnh.show()
```

```python
import cv2
import matplotlib.pyplot as LeDucAnh
img = cv2.imread('hinh1.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
face_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
found = face_data.detectMultiScale(img_gray, scaleFactor=1.3, minNeighbors=5)
amount_found = len(found)
if amount_found != 0:
    for (x, y, width, height) in found:
        cv2.rectangle(img_rgb, (x, y),
                      (x + width, y + height),
                      (0, 255, 0), 3)
LeDucAnh.imshow(img_rgb)
LeDucAnh.title("Face Detection")
LeDucAnh.axis('off')
LeDucAnh.show()
```

```python
import cv2
import matplotlib.pyplot as TenSV

img = cv2.imread('hinh1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

eyes = eye_cascade.detectMultiScale(gray, 1.3, 5)

for (x,y,w,h) in eyes:
    cv2.rectangle(rgb, (x,y), (x+w,y+h), (255,255,0), 2)

TenSV.imshow(rgb)
TenSV.title("Eye Detection")
TenSV.axis('off')
TenSV.show()
```

```python
import cv2
import matplotlib.pyplot as LeDucAnh
img = cv2.imread('hinh1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.equalizeHist(gray)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')
faces = face_cascade.detectMultiScale(gray, 1.2, 4)
for (x,y,w,h) in faces:
    face_gray = gray[y:y+h, x:x+w]
    face_color = rgb[y:y+h, x:x+w]
    smiles = smile_cascade.detectMultiScale(face_gray, 1.3, 10)
    for (sx,sy,sw,sh) in smiles:
        cv2.rectangle(face_color, (sx,sy), (sx+sw,sy+sh), (255,255,0), 2)
LeDucAnh.imshow(rgb)
LeDucAnh.title("Smile Detection (Improved)")
LeDucAnh.axis('off')
LeDucAnh.show()
```

```python
import cv2
import matplotlib.pyplot as LeDucAnh
img = cv2.imread('hinh1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
nose_cascade = cv2.CascadeClassifier('haarcascade_mcs_nose.xml')
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    face_gray = gray[y:y+h, x:x+w]
    face_color = rgb[y:y+h, x:x+w]
    noses = nose_cascade.detectMultiScale(face_gray, 1.3, 5)
    for (nx,ny,nw,nh) in noses:
        cv2.rectangle(face_color, (nx,ny), (nx+nw,ny+nh), (255,0,255), 2)
LeDucAnh.imshow(rgb)
LeDucAnh.title("Nose Detection")
LeDucAnh.axis('off')
LeDucAnh.show()
```
