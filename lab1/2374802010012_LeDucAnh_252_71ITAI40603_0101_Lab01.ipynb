```python
import cv2 as cv
import matplotlib.pyplot as LeDucAnh
```

```python
img = cv.imread('hinh1.jpg')
```

```python
LeDucAnh.imshow(img)
LeDucAnh.title("Anh goc")
LeDucAnh.show()
```

```python
print(img.shape)
```

Ảnh sẽ bị sai màu vì OpenCV đọc ảnh theo BGR, còn matplotlib hiển thị RGB

```python
img_rgb = img[:,:,::-1]

LeDucAnh.imshow(img_rgb)
LeDucAnh.title("Anh dung mau")
LeDucAnh.show()
```
