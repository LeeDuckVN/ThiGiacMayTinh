```python
import cv2
import numpy as np
import matplotlib.pyplot as LeDucAnh
```

```python
img = cv2.imread('hinh1.png')
img = img[:,:,::-1]

LeDucAnh.imshow(img)
LeDucAnh.title("Input Image")
LeDucAnh.show()
```

```python
img = cv2.imread('hinh1.png',0)

img_float32 = np.float32(img)

dft = cv2.dft(img_float32, flags=cv2.DFT_COMPLEX_OUTPUT)

dft_shift = np.fft.fftshift(dft)

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))

LeDucAnh.figure(figsize=(10,5))

LeDucAnh.subplot(1,2,1)
LeDucAnh.imshow(img,cmap='gray')
LeDucAnh.title('Input Image')

LeDucAnh.subplot(1,2,2)
LeDucAnh.imshow(magnitude_spectrum,cmap='gray')
LeDucAnh.title('FFT Magnitude Spectrum')

LeDucAnh.show()
```

```python
img = cv2.imread('hinh2.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

dst = cv2.cornerHarris(gray,2,3,0.04)

dst = cv2.dilate(dst,None)

img[dst > 0.01 * dst.max()] = [0,0,255]

LeDucAnh.figure(figsize=(8,8))
LeDucAnh.imshow(img[:,:,::-1])
LeDucAnh.title("Harris Corner Detection")
LeDucAnh.show()
```

```python
import numpy as np
from skimage.data import gravel
from skimage.filters import difference_of_gaussians

image = gravel()

filtered_image = difference_of_gaussians(image, 1, 12)

LeDucAnh.figure(figsize=(10,5))

LeDucAnh.subplot(1,2,1)
LeDucAnh.imshow(image,cmap='gray')
LeDucAnh.title("Original Image")

LeDucAnh.subplot(1,2,2)
LeDucAnh.imshow(filtered_image,cmap='gray')
LeDucAnh.title("Filtered Image")

LeDucAnh.show()
```

```python
img = cv2.imread('hinh1.png',0)

ddepth = cv2.CV_16S

grad_x = cv2.Sobel(src=img, ddepth=ddepth, dx=1, dy=0, ksize=3)

abs_grad_x = cv2.convertScaleAbs(grad_x)

LeDucAnh.figure(figsize=(8,8))
LeDucAnh.imshow(abs_grad_x,cmap='gray')
LeDucAnh.title("Sobel Edge X")
LeDucAnh.show()
```

```python
grad_y = cv2.Sobel(src=img, ddepth=ddepth, dx=0, dy=1, ksize=3)

abs_grad_y = cv2.convertScaleAbs(grad_y)

LeDucAnh.figure(figsize=(8,8))
LeDucAnh.imshow(abs_grad_y,cmap='gray')
LeDucAnh.title("Sobel Edge Y")
LeDucAnh.show()
```

```python
edge = cv2.addWeighted(abs_grad_x,0.5,abs_grad_y,0.5,0)

LeDucAnh.figure(figsize=(8,8))
LeDucAnh.imshow(edge,cmap='gray')
LeDucAnh.title("Combined Edge Detection")
LeDucAnh.show()
```
