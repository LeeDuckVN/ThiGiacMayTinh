```python
import cv2
import matplotlib.pyplot as plt
```

Sử dụng hình ảnh cat.png để đọc ảnh và hiển thị ảnh ban đầu.


```python
# Đọc ảnh cat.png
img = cv2.imread('cat.png')
img = img[:, :, ::-1]   # Chuyển từ BGR sang RGB
```

# 1.Sử dụng boxFilter: cv2.blur() hoặc cv2.boxFilter() để làm mờ ảnh (sử dụng nhiều thông số filter khác nhau), biểu diễn ảnh gốc và ảnh làm mờ cùng nhau để kiểm chứng.  


```python
anh_loc_3 = cv2.blur(img, (3, 3))      # Lọc Box kích thước 3x3
anh_loc_7 = cv2.blur(img, (7, 7))      # Lọc Box kích thước 7x7
anh_loc_11 = cv2.blur(img, (11, 11))   # Lọc Box kích thước 11x11

# Hiển thị ảnh gốc và ảnh sau khi làm mờ
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(img)
plt.title("Ảnh gốc")

plt.subplot(1,4,2)
plt.imshow(anh_loc_3)
plt.title("Lọc Box 3x3")

plt.subplot(1,4,3)
plt.imshow(anh_loc_7)
plt.title("Lọc Box 7x7")

plt.subplot(1,4,4)
plt.imshow(anh_loc_11)
plt.title("Lọc Box 11x11")

plt.show()
```

# 2.Sử dụng Gaussian Filter # Lưu ý: kích thước bộ lọc là số lẻ.

```python
anh_loc_gaussian_3 = cv2.GaussianBlur(img, (3,3), 0)   # Lọc Gaussian 3x3
anh_loc_gaussian_5 = cv2.GaussianBlur(img, (5,5), 0)   # Lọc Gaussian 5x5
anh_loc_gaussian_7 = cv2.GaussianBlur(img, (7,7), 0)   # Lọc Gaussian 7x7

# Hiển thị ảnh gốc và ảnh sau khi lọc Gaussian
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(img)
plt.title("Ảnh gốc")

plt.subplot(1,4,2)
plt.imshow(anh_loc_gaussian_3)
plt.title("Lọc Gaussian 3x3")

plt.subplot(1,4,3)
plt.imshow(anh_loc_gaussian_5)
plt.title("Lọc Gaussian 5x5")

plt.subplot(1,4,4)
plt.imshow(anh_loc_gaussian_7)
plt.title("Lọc Gaussian 7x7")

plt.show()
```

# 3.Sử dụng Median Filter 

```python
# Sử dụng bộ lọc Median để làm mờ ảnh

anh_loc_median_3 = cv2.medianBlur(img, 3)   # Lọc Median với kích thước 3
anh_loc_median_5 = cv2.medianBlur(img, 5)   # Lọc Median với kích thước 5
anh_loc_median_7 = cv2.medianBlur(img, 7)   # Lọc Median với kích thước 7

# Hiển thị ảnh gốc và ảnh sau khi lọc Median
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(img)
plt.title("Ảnh gốc")

plt.subplot(1,4,2)
plt.imshow(anh_loc_median_3)
plt.title("Lọc Median 3x3")

plt.subplot(1,4,3)
plt.imshow(anh_loc_median_5)
plt.title("Lọc Median 5x5")

plt.subplot(1,4,4)
plt.imshow(anh_loc_median_7)
plt.title("Lọc Median 7x7")

plt.show()
```

# 7.Bài tập trên lớp lý thuyết: FILTER bằng các phép toán số học và logic
Given two images as follow (with different sizes: 193x157 and 194x160) 

```python
import cv2
import matplotlib.pyplot as plt

# Đọc hai ảnh (đúng tên file trong thư mục)
img1 = cv2.imread('img1.png')
img2 = cv2.imread('img2.png')

# Kiểm tra đọc ảnh
if img1 is None or img2 is None:
    raise ValueError("Không đọc được ảnh img1.png hoặc img2.png")

# Cắt hai ảnh về cùng kích thước 190x155 (đúng như đề)
img1 = img1[:155, 1:191]   # 155 hàng, 190 cột
img2 = img2[:155, :190]

# Chuyển từ BGR sang RGB để hiển thị
img1 = img1[:, :, ::-1]
img2 = img2[:, :, ::-1]

# Hiển thị ảnh sau khi cắt
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img1)
plt.title("Ảnh img1 sau khi cắt")

plt.subplot(1,2,2)
plt.imshow(img2)
plt.title("Ảnh img2 sau khi cắt")
plt.show()
```

```python
# Tạo ảnh âm bản bằng phép toán logic
inv_img1 = cv2.bitwise_not(img1)
inv_img2 = cv2.bitwise_not(img2)

# Hiển thị ảnh âm bản
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(inv_img1)
plt.title("img1 âm bản")

plt.subplot(1,2,2)
plt.imshow(inv_img2)
plt.title("img2 âm bản")
plt.show()
```

```python
# Dùng phép toán AND để giữ lại hình tròn
anh_ball = cv2.bitwise_and(inv_img1, inv_img2)

# Hiển thị kết quả
plt.imshow(anh_ball)
plt.title("Ảnh chỉ còn hình tròn (ball)")
plt.show()
```

Các vạch trắng xuất hiện do nhiễu hoặc sai khác cường độ điểm ảnh sau các phép toán logic.

## Một số bài tập mở rộng

1. Sử dụng Backward, Forward, Central Gradient để lọc ảnh

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
img = cv2.imread('cat.png')
img = img[:, :, ::-1]
anh_xam = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Forward Gradient
forward_x = np.zeros_like(anh_xam)
forward_x[:, :-1] = anh_xam[:, 1:] - anh_xam[:, :-1]

# Backward Gradient
backward_x = np.zeros_like(anh_xam)
backward_x[:, 1:] = anh_xam[:, 1:] - anh_xam[:, :-1]

# Central Gradient
central_x = np.zeros_like(anh_xam)
central_x[:, 1:-1] = anh_xam[:, 2:] - anh_xam[:, :-2]

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(forward_x, cmap='gray'); plt.title("Forward Gradient")
plt.subplot(1,3,2); plt.imshow(backward_x, cmap='gray'); plt.title("Backward Gradient")
plt.subplot(1,3,3); plt.imshow(central_x, cmap='gray'); plt.title("Central Gradient")
plt.show()
```

2. Sử dụng Finite Difference Filter để lọc ảnh nhiễu

```python
# Kernel Finite Difference
kernel_fd_x = np.array([[-1, 1]])
kernel_fd_y = np.array([[-1], [1]])

fd_x = cv2.filter2D(anh_xam, -1, kernel_fd_x)
fd_y = cv2.filter2D(anh_xam, -1, kernel_fd_y)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(fd_x, cmap='gray'); plt.title("Finite Difference X")
plt.subplot(1,2,2); plt.imshow(fd_y, cmap='gray'); plt.title("Finite Difference Y")
plt.show()
```

3. Sử dụng Gaussian Filter để lọc ảnh

```python
anh_gaussian = cv2.GaussianBlur(anh_xam, (5,5), 0)

plt.imshow(anh_gaussian, cmap='gray')
plt.title("Ảnh sau khi lọc Gaussian")
plt.show()
```

4. Sử dụng Sobel detector để xác định biên của ảnh

```python
sobel_x = cv2.Sobel(anh_xam, cv2.CV_64F, 1, 0)
sobel_y = cv2.Sobel(anh_xam, cv2.CV_64F, 0, 1)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(sobel_x, cmap='gray'); plt.title("Sobel theo trục X")
plt.subplot(1,2,2); plt.imshow(sobel_y, cmap='gray'); plt.title("Sobel theo trục Y")
plt.show()
```

5. Sử dụng Canny edge detector để xác định các cạnh góc ảnh

```python
anh_canny = cv2.Canny(anh_xam, 100, 200)

plt.imshow(anh_canny, cmap='gray')
plt.title("Phát hiện cạnh bằng Canny")
plt.show()
```

6. Sử dụng Hough Transform để xác định các đường thẳng trong ảnh

```python
# Phát hiện cạnh trước bằng Canny
edges = cv2.Canny(anh_xam, 100, 200)

# Hough Transform
lines = cv2.HoughLines(edges, 1, np.pi/180, 150)

# Vẽ các đường thẳng tìm được
anh_hough = img.copy()
if lines is not None:
    for i in range(min(len(lines), 5)):
        rho, theta = lines[i][0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(anh_hough, (x1,y1), (x2,y2), (255,0,0), 2)

plt.imshow(anh_hough)
plt.title("Phát hiện đường thẳng bằng Hough Transform")
plt.show()
```
